Pandas' data structures are called “dataframes”. Let's create one from scratch:

```python
table = [
   ['UK', 2678454886796.7],    # 1st row
   ['USA', 16768100000000.0],  # 2nd row
   ['China', 9240270452047.0], # and so on...
   ['Brazil', 2245673032353.8],
   ['South Africa', 366057913367.1]
]
headings = ['Country', 'GDP (US$)']
```

```python
import pandas
from pandas import read_csv
df_val = read_csv('my_predictions_val.csv')
df_val.info()
```

```python
def CalculError(row):
    v = row['SeriousDlqin2yrs']
    w = row['SeriousDlqin2yrs.1']
    if v == 0 :
        if w == 0 :
            return 'TN'
        else :
            return 'FP'

    else :
        if w == 0 :
            return 'FN'
        else :
            return 'TP'
        
```

```python
df_val['error']= df_val.apply (CalculError, axis = 1)
df_val.head (10)
resultat = df_val.groupby('error')['field1'].nunique()
df2 = pandas.DataFrame({'Prediction 0' : [resultat[2],resultat[0]], 'Prediction 1':[resultat[1],resultat[3]]},index = ['Actual 0', 'Actual 1'] )
print(df2)


```
